In order to work with python-ggseg, the data should be prepared as a dictionary where each item is one region of a given atlas assigned with some numeric value. The current version includes three atlases: the Desikan-Killiany (DK) atlas , the Johns Hopkins University (JHU) white-matter atlas and the FreeSurfer aseg atlas.

```python
import glob

atlas_path = (
    "/home/mwagner/bin/anaconda3/envs/ggseg/lib/python3.8/site-packages/ggseg/data/"
)

ctx = [f.split("/")[-1] for f in glob.glob(atlas_path + "dk/*")]
subcort = [f.split("/")[-1] for f in glob.glob(atlas_path + "aseg/*")]

print(f"N CORTEX ROIS:\t{len(ctx)}")
print(f"N SUBCORT ROIS:\t{len(subcort)}")
print()
print(f"TOTAL:\t\t{len(subcort)+len(ctx)}")
```

```python
data = {
    "bankssts_left": 1.1,
    "caudalanteriorcingulate_left": 1.0,
    "caudalmiddlefrontal_left": 2.6,
    "cuneus_left": 2.6,
    "entorhinal_left": 0.6,
}
```

```python
import ggseg

ggseg.plot_dk(
    data,
    cmap="Spectral",
    figsize=(15, 15),
    background="k",
    edgecolor="w",
    bordercolor="gray",
    ylabel="Cortical thickness (mm)",
    title="Title of the figure",
)
```
