```python
import os

def is_subdirectory(parent, child):
    """
    Check if 'child' is a subdirectory of 'parent'.
    """
    parent = os.path.abspath(parent)
    child = os.path.abspath(child)
    return child.startswith(parent)


```

```python
# Example usage:
parent_directory = "/snl/abcd/raw-data/smri/year2/"
child_directory = "/sub-NDARINV00U4FTRU"

if is_subdirectory(parent_directory, child_directory):
    print(f"{child_directory} is a subdirectory of {parent_directory}")
else:
    print(f"{child_directory} is not a subdirectory of {parent_directory}")
```
